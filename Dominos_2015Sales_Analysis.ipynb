```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.express.colors as px_colors
```

```python
sb=pd.read_csv('/content/pizza_sales.csv')
sb
```

```python
sb.info()
```

```python
# Checking duplicates
sb.duplicated().sum()
```

```python
# Converting unit_price and total_price usd in inr
sb.unit_price=sb.unit_price*86.50
sb.total_price=sb.total_price*86.50
```

```python
# Rounding unit_price and total_price
sb['unit_price']=sb['unit_price'].round()
sb['total_price']=sb['total_price'].round()
```

```python
# Converting data type of unit_price and total_price
sb['unit_price']=sb['unit_price'].astype(int)
sb['total_price']=sb['total_price'].astype(int)
sb.info()
```

```python
# Converting order_date data type
sb['order_date']=pd.to_datetime(sb['order_date'])
```

```python
sb.info()
```

#EDA

#Top 10 pizza ingredients people preferred the most

```python
# copying pizza_ingredients to new column
sb['pizza_ingr_combination'] = sb['pizza_ingredients']
```

```python
# Here before exploding we are splittling ingredients to transform each ingredient into an individual element in a list
sb['pizza_ingredients'] = sb['pizza_ingredients'].str.split(',')
# Exploding the list of ingredients into individual rows
sb_exploded = sb.explode('pizza_ingredients')
```

```python
# Total counts of pizza ingredients
ingredient_counts = sb_exploded['pizza_ingredients'].value_counts()
print(ingredient_counts)
```

```python
#Top 10 pizza ingredients people preferred the most
top10_Ing=ingredient_counts.head(10).reset_index()
top10_Ing
```

```python
px.bar(top10_Ing, x='pizza_ingredients',y='count', color='pizza_ingredients', color_discrete_sequence=px.colors.qualitative.Pastel1,
       labels={'x':'Pizza Ingredients','y':'Count'},
       title='Top 10 pizza ingredients people preferred the most')
```

#Distribution of Monthly Orders in 2015

```python
# Extracting month from date column
sb['month'] = sb['order_date'].dt.month
# Getting month_name in new column from month column
sb['month_name']=sb['order_date'].dt.month_name()
sb.sample()
```

```python
# total order in each month
monthly_order=sb['month_name'].value_counts().reset_index()
monthly_order
```

```python
px.pie(monthly_order,names='month_name', values='count', hole=0.4,
       labels={'month_name':'Month','count':'Monthly Orders'},
       title='Distribution of Monthly Orders in 2015')
```

#Top 5 pizza ingredients combination people preferred the most

```python
top5_PIC=sb['pizza_ingr_combination'].value_counts().head(5).reset_index()
top5_PIC
```

```python
fig=px.pie(top5_PIC, names='pizza_ingr_combination',values='count',
       labels={'x':'Pizza Ingredients Combination','y':'Count'},
       title='Top 5 pizza ingredients combination people preferred the most',)

# pull parameter is not directly supported in the px.pie() used for advanced attributes
fig.update_traces(pull=[0.2, 0, 0, 0, 0])
fig.show()
```

#Total Sales amount by Pizza Category

```python
# Grouping data by pizza_category and calculating total_price
total_price_by_category=sb.groupby('pizza_category')['total_price'].sum().reset_index()
total_price_by_category
```

```python
px.bar(total_price_by_category,x='pizza_category',y='total_price', color='pizza_category', color_discrete_sequence=px.colors.qualitative.Set2,
       labels={'pizza_category':'Pizza Category','total_price':'Total Sales'},
       title='Total Sales amount by Pizza Category')
```

#Total Sales Over Time in 2015

```python
# Grouping data by order_date and calculating total_price
orderdate=sb.groupby('order_date')['total_price'].sum().reset_index()
orderdate
```

```python
px.line(orderdate, x='order_date', y='total_price',
        labels={'order_date': 'Order Date', 'total_price': 'Total Sales'},
        title='Total Sales Over Time in 2015')
```

```python
top5_Pizza=sb['pizza_name'].value_counts().head(10).reset_index()
top5_Pizza
```
